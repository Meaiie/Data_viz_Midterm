## 1. ความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

```python
Tidy_data.isnull().sum()
```

```python
# สร้างคอลัมน์ year เพิ่ม
Tidy_data['year'] = Tidy_data['month'].str[-4:]
Tidy_data
```

```python
# Group the data by 'Station' and 'Year', then calculate the mean of 'Chlo_a' for each group
grouped_data = Tidy_data.groupby(['station', 'year'])['Chlo_a'].mean().reset_index()

# Pivot the table to have years as columns and stations as rows
pivot_table = grouped_data.pivot(index='station', columns='year', values='Chlo_a')

# Calculate the difference between the mean Chlo_a values for 2018 and 2019
pivot_table['Difference'] = pivot_table['2019'] - pivot_table['2018']

for idx, value in pivot_table['Difference'].items():
  if value > 0:
     pivot_table.loc[idx, 'Difference'] = value
  else:
     pivot_table.loc[idx, 'Difference'] = -value

pivot_table
```

## หาว่าคนส่วนใหญ่เป็นมะเร็งชนิดไหน

```python
import re

def find_all_cancer_types(comment):
    # ใช้ regex ในการค้นหาคำที่ตามหลัง 'มะเร็ง' ตามด้วยคำที่ไม่ใช่ช่องว่างหรือจุด
    matches = re.findall(r'มะเร็ง(\S+)', comment)

    # คืนค่าลิสต์ของชื่อมะเร็งที่พบ ถ้าไม่เจอจะได้ลิสต์ว่าง
    return matches if matches else None

# ใช้ฟังก์ชันกับคอลัมน์ 'comments' เพื่อสร้างคอลัมน์ 'cancer_type'
data_dr_filtered['cancer_type'] = data_dr_filtered['comments'].apply(find_all_cancer_types)
data_cancer = data_dr_filtered.copy()

data_cancer
```

```python
# ข้อมูลชนิดมะเร็งทั้งหมด
cancer_data = {
    'ชนิดมะเร็ง (ภาษาไทย)': [
        'เต้านม','จมูก','ปอด', 'ตับ', 'ตับอ่อน', 'กระเพาะอาหาร',
        'ลำไส้ใหญ่', 'กระเพาะปัสสาวะ', 'มดลูก', 'รังไข่',
        'ต่อมลูกหมาก', 'เม็ดเลือดขาว', 'ต่อมน้ำเหลือง', 'ช่องปาก',
        'หลอดอาหาร', 'สมอง', 'ผิวหนัง', 'ปากมดลูก',
        'ต่อมไทรอยด์', 'ไขกระดูก', 'ต่อมหมวกไต', 'ถุงน้ำดี',
        'ปอดชนิดเซลล์ขนาดเล็ก', 'ปอดชนิดเซลล์ไม่เล็ก', 'อัณฑะ',
        'ถุงน้ำดี', 'ปากมดลูก', 'หลอดอาหาร', 'ต่อมลูกหมาก','กระดูก','ลำไส้','ลำใส้'
        ,'ลำคอ','ไทรอยด์','ท่อน้ำดี','ไต','ไทรอย','ต่อมน้ำลาย','โพรงจมูก','กระเพาะปัสสาวะ'
        ,'ลิ้น','กะเพาะปัสสาวะ','กระเพาะ','กล่องเสียง','ต่อมทอนซิล','เนื้อเยื่อบุผิว'
        ,'ปาก','กรามช้าง','ท่อน้ําดี','ต่อมไทรมัส','ซาร์โคม่า','ลำใส่','เม็ดเลือด'
    ]
}

cancer_df = pd.DataFrame(cancer_data)
cancer_df
```

```python
# ฟังก์ชันค้นหาชนิดมะเร็งที่ประกอบด้วยคำในตาราง
def find_cancer_types_in_comment(cancer_list):
    if not isinstance(cancer_list, list):  # ตรวจสอบว่าค่าเป็นลิสต์
        return None
    matched_cancers = []
    for cancer in cancer_list:
        for keyword in cancer_df['ชนิดมะเร็ง (ภาษาไทย)']:
            if keyword in cancer:
                matched_cancers.append(keyword)

    return matched_cancers if matched_cancers else None

# ฟิลเตอร์และสร้างคอลัมน์ใหม่ 'matched_cancers'
data_cancer['matched_cancers'] = data_cancer['cancer_type'].apply(find_cancer_types_in_comment)

# แสดง DataFrame พร้อมคอลัมน์ใหม่
data_cancer
```

```python
data_cancer[data_cancer['matched_cancers'].isnull()]
```

```python
# prompt: เลือก 1 อันจากลิสต์ในคอลัมน์ matched_cancers มา

def select_one_cancer(cancer_list):
  if cancer_list:
    return cancer_list[0]  # เลือกตัวแรกในลิสต์
  else:
    return None

# ใช้ฟังก์ชันกับคอลัมน์ 'matched_cancers' เพื่อสร้างคอลัมน์ 'selected_cancer'
data_cancer['selected_cancer'] = data_cancer['matched_cancers'].apply(select_one_cancer)

# แสดง DataFrame พร้อมคอลัมน์ใหม่
data_cancer
```

```python
def replace_cancer_names(cancer_name):
  if cancer_name == 'ลำใส้' or cancer_name == 'ลำใส่':
    return 'ลำไส้'
  if cancer_name == 'ไทรอย' or cancer_name == 'ไทรอยด์':
    return 'ต่อมไทรอยด์'
  if cancer_name == 'กะเพาะปัสสาวะ':
    return 'กระเพาะปัสสาวะ'
  if cancer_name == 'กระเพาะ':
    return 'กระเพาะอาหาร'
  if cancer_name == 'ท่อน้ําดี':
    return 'ท่อน้ำดี'
  if cancer_name == 'ปาก':
    return 'ช่องปาก'
  else:
    return cancer_name

data_cancer['selected_cancer'] = data_cancer['selected_cancer'].apply(replace_cancer_names)

# แสดง DataFrame พร้อมค่าที่ถูกแทนที่
data_cancer
```

```python
import re

def find_cancer_phase(comment):
    # ค้นหาข้อความที่เริ่มจาก 'ระยะ' ตามด้วยเลข
    match = re.search(r'ระยะ[^0-9]*\d+', comment)
    if match:
        return match.group(0)

    # ถ้าไม่พบเลข ค้นหาข้อความที่เริ่มจาก 'ระยะ' ตามด้วยคำเฉพาะ
    match = re.search(r'ระยะ(?:ที่|[^\d\s]*)?(?:แรก|หนึ่ง|สอง|สาม|สี่|สุดท้าย|ลุกลาม|เริ่ม|ก่อน)', comment)

    return match.group(0) if match else None

# ใช้ฟังก์ชันกับคอลัมน์ 'comments' เพื่อสร้างคอลัมน์ 'cancer_phase'
data_cancer_phase['cancer_phase'] = data_cancer_phase['comments'].apply(find_cancer_phase)

data_cancer_phase
```
